## Question 3 ##

```python
#Data Set 1
data = {
    'Player': ['Grayson Allen', 'Harrison Barnes', 'Josh Giddey', 'Elijah Hughes', 
               'DeMar DeRozan', 'Rudy Gobert', 'Giannis Atetokounmpo', 
               'Mike Conley', 'Anthony Davis'],
    '3PT Percentage': [40.7, 39.4, 26.4, 22.2, 36.5, 0.0, 30.0, 41.3, 18.6],
    'Free Throw Percentage': [86.5, 82.7, 70.9, 75.0, 87.7, 68.9, 72.0, 79.6, 71.1],
    'Defensive Rebound Percentage': [10.1, 13.2, 17.1, 11.7, 12.4, 30.7, 25.9, 8.1, 20.4]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Normalize the data
scaler_x = MinMaxScaler()
scaler_ftp = MinMaxScaler()
scaler_y = MinMaxScaler()
ftp = df[['Free Throw Percentage']].values
ftp = scaler_ftp.fit_transform(ftp)
x = df[['Free Throw Percentage', 'Defensive Rebound Percentage']].values
x = scaler_x.fit_transform(x)
y = df[['3PT Percentage']].values
y = scaler_y.fit_transform(y)
print(x)
print(y)
```

```python
#Create the tree
tree_test_ftp = tree.DecisionTreeRegressor(random_state=0, min_samples_split=4)
tree_test_ftp.fit(ftp,y)
# Plot the decision boundary
tree.plot_tree(tree_test_ftp, feature_names=['Free Throw Percentage'], filled=True)
```

```python
#Create the tree
tree_test = tree.DecisionTreeRegressor(random_state=0, min_samples_split=4)
tree_test.fit(x,y)
# Plot the decision boundary
tree.plot_tree(tree_test, feature_names=['Free Throw Percentage', 'Defensive Rebound Percentage'], filled=True)
```

```python
ax = plt.gca()
plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)
DecisionBoundaryDisplay.from_estimator(
    tree_test,
    x,
    cmap=plt.cm.RdYlBu,
    response_method="predict",
    ax=ax,
)

# Plot the training points
# for i, color in zip(range(n_classes), plot_colors):
#     idx = np.where(y == i)
plt.scatter(
    x[:, 0],
    x[:, 1],
    cmap=plt.cm.RdYlBu,
    edgecolor="black",
    s=15,
)

plt.suptitle("Decision surface of decision trees trained on pairs of features")
plt.legend(loc="lower right", borderpad=0, handletextpad=0)
plt.xlim([-.1, 1.1])
plt.ylim([-.1, 1.1])
# _ = plt.axis("tight")
plt.xlabel('Free Throw Percentage')
plt.ylabel('Defensive Rebound Percentage')
plt.show()
```

```python


data = [
    ("Let This House Be Built unto My Name", "Elder Bednar", [10.43, 5.72]),
    ("We Will Prove Them Herewith", "Elder Bednar", [5.71, 4.14]),
    ("With the Power of God in Great Glory", "Elder Bednar", [9.18, 6.18]),
    ("But We Heeded Them Not", "Elder Bednar", [8.82, 5.15]),
    ("Embrace the Future with Faith", "President Nelson", [5.23, 2.47]),
    ("Christ is Risen; Faith in Him Will Move Mountains", "President Nelson", [5.66, 3.14]),
    ("The Temple and Your Spiritual Foundation", "President Nelson", [6.74, 2.73]),
    ("The Power of Spiritual Momentum", "President Nelson", [5.28, 2.91])
]

df = pd.DataFrame(data, columns=["Title", "Speaker", "xT"])

print(df)
```

```python
scaler_x = MinMaxScaler()
df[['xT1', 'xT2']] = pd.DataFrame(df['xT'].tolist(), index=df.index)
x = df[['xT1', 'xT2']].values
x = scaler_x.fit_transform(x)
print(x)
```

```python
tree_the = tree.DecisionTreeClassifier(random_state=0, min_samples_split=4, criterion="entropy")
y = df[['Speaker']]
x_the = x[:,0]
print(x_the)
tree_the.fit(x_the.reshape(-1,1),y)
tree.plot_tree(tree_the,class_names = ["Bednar", "Nelson"], feature_names = ["the"])

```

```python
tree_both = tree.DecisionTreeClassifier(random_state=0, min_samples_split=4, criterion="entropy")
tree_both.fit(x,y)
tree.plot_tree(tree_both, filled=True, class_names = ["Bednar", "Nelson"], feature_names = ["the", "of"])

```

```python
ax = plt.gca()
plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)
DecisionBoundaryDisplay.from_estimator(
    tree_both,
    x,
    cmap=plt.cm.RdYlBu,
    response_method="predict",
    ax=ax,
    class_names = ["Bednar", "Nelson"],
    feature_names = ["the", "of"],


)
scatter = plt.scatter(    
    x[:,0],
    x[:,1],
)
plt.xlim([-.1, 1.1])
plt.ylim([-.1, 1.1])
plt.xlabel("the")
plt.ylabel("of")
plt.show()
```
